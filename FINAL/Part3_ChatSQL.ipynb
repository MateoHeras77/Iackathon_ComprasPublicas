```python
import pandas as pd
import sqlite3

# Supongamos que tienes DataFrames df_2015, df_2016, ..., df_2022

# Crear una lista para almacenar los DataFrames de muestra
sampled_dataframes = []

# Itera a través de los DataFrames y obtén una muestra aleatoria de 1000 datos
for year in range(2015, 2023):
    df_name = f'df_{year}'
    sampled_df = globals()[df_name].sample(n=30000, random_state=42, replace=True)
    sampled_dataframes.append(sampled_df)

# Concatena todos los DataFrames de muestra en uno solo
combined_df = pd.concat(sampled_dataframes, ignore_index=True)

# Elimina filas con valores NaN
combined_df.dropna(inplace=True)

# Crea una base de datos SQLite 3
conn = sqlite3.connect('datos.db')

# Guarda el DataFrame en la base de datos con el nombre 'sampled_data'
combined_df.to_sql('sampled_data', conn, if_exists='replace', index=False)

# Cierra la conexión a la base de datos
conn.close()
```
